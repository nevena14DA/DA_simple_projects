## Q. 4) Question ( mapping + data-type casting ) - What is the mean stop_duration ?

```python
police_df.head()
```

```python
police_df['stop_duration'].value_counts()
```

```python
## stop_duration needs to be converted to int in order to do mean
## 0-15:7.5 means convert 0-15 minuts to average of 7.5 minutes
police_df['stop_duration']=police_df['stop_duration'].map({'0-15 Min':7.5 ,'16-30 Min':24, '30+ Min':45})
```

```python
police_df.head()
```

```python
police_df['stop_duration'].mean()
```
